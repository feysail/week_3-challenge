```python
contingency_table_provinces = pd.crosstab(df['PostalCode'], df['Province'])
chi2_provinces, p_provinces, _, _ = stats.chi2_contingency(contingency_table_provinces)

chi2_provinces, p_provinces, _, _
```

```python
# 4. Risk differences between women and men
risk_women = df[df['Gender'] == 'Female']['TotalPremium']
risk_men = df[df['Gender'] == 'Male']['TotalPremium']
t_stat_gender, p_gender = stats.ttest_ind(risk_women, risk_men)

print(t_stat_gender, p_gender)
```
